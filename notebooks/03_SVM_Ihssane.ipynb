```python
import shutil, os

REPO = '/content/drive/MyDrive/flipkart_sentiments_pfa'
colab_dir = '/content/drive/MyDrive/Colab Notebooks'

# Déplacer les 2 fichiers utiles dans le repo
fichiers = {
    'Copie de 02_Preprocessing_Ihssane.ipynb': f'{REPO}/notebooks/02_Preprocessing_Ihssane.ipynb',
    '03_Models_SVM_Ihssane.ipynb':    f'{REPO}/notebooks/03_Models_SVM_Ihssane.ipynb',
}

for src_name, dst_path in fichiers.items():
    src_path = f'{colab_dir}/{src_name}'
    if os.path.exists(src_path):
        shutil.copy(src_path, dst_path)
        print(f"✅ Copié : {src_name}")
    else:
        print(f"❌ Introuvable : {src_name} — vérifie le nom exact")

# Vérifier le résultat
print("\n📂 notebooks/ contient maintenant :")
for f in os.listdir(f'{REPO}/notebooks'):
    print(" ", f)
```
